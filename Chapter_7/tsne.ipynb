```python
import numpy as np
np.random.seed(0)
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
```

```python
dataset = load_digits()

X: np.ndarray = dataset.data
y: np.ndarray = dataset.target

n_samples, n_features = X.shape
```

```python
n_components = 2
pca = PCA(n_components=n_components, copy=True)

x_reduced = pca.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_reduced, y, test_size=0.3)
```

```python
colors = ["red", "blue", "orange", "yellow", "cyan", "black", "green", "purple", "pink", "gray"]

plt.figure(figsize=(8, 8))
for index, point in enumerate(x_test):
    plt.scatter(point[0], point[1], color=colors[y_test[index]])
plt.xlabel("svd feature 1")
plt.ylabel("svd feature 2")
plt.show()
```

```python
n_components = 2
tsne = TSNE(n_components=n_components, n_iter=1500)
x_reduced = tsne.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_reduced, y, test_size=0.3)
```

```python
plt.figure(figsize=(8, 8))
for index, point in enumerate(x_test):
    color = colors[y_test[index]]
    plt.scatter(point[0], point[1], color=color)
plt.xlabel("tsne feature 1")
plt.ylabel("tsne feature 2")
plt.show()
```
